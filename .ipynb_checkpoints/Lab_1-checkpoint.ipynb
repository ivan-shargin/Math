```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

 В качестве функции возьмем координату от времени при затухающих колебаниях

```python
A0 = 1 
attenuation = 0.1
att = attenuation
period = 10
w = 2 * math.pi / period
```

```python
def func(t, A0, attenuation, w):
    return A0 * np.exp(-attenuation * t) * np.sin(w * t)
```

```python
t = np.linspace(0, period, 100)
f = func(t, A0, attenuation, w)
plt.plot(t, f)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Изначальная функция')
plt.grid()
```

Функция "real_dif_1" будет рассчитывать первую производную по точной формуле.

```python
def real_dif_1(t, A0, attenuatuion, w):
    return A0 * np.exp(-attenuation * t) * (w * np.cos(w * t) - attenuation * np.sin(w * t))
    
```

## Первая производная с первым порядком точности

Функция dif_11(func_cell, t) рассчитывает первую производную сеточной функции (func_cell) на отрезке t(list t) c первым порядком точности

```python
def dif_11(func_cell, t):
    dif_cell = [0] * len(func_cell) 
    h = t[1] - t[0]
    
    dif_cell[-1] = (func_cell[-1] - func_cell[-2]) / h
    
    for i in range (len(func_cell) - 1):
        dif_cell[i] = (func_cell[i+1] - func_cell[i]) / h
        
    return dif_cell
    
```

```python
t = np.linspace(0, period, 30)
f = func(t, A0, attenuation, w)
d11 = dif_11(f, t)
plt.plot(t, d11, label = 'расчетная производная')

# err = np.abs(dif_11 - real_dif_1(t, A0, attenuation, w))

plt.plot(t, real_dif_1(t, A0, attenuation, w), label='точная производная')
plt.legend()
```

```python
N_min = 100
N_max = 1000
Er_array_11 = []
h_arr = []
for N in range (N_min, N_max):
    t = np.linspace(0, period, N)
    h = period / (N - 1)
    h_arr.append(h)
    func_cell = func(t, A0, attenuation, w)
    D_11 = dif_11(func_cell, t)
    err = max(np.abs(D_11 - real_dif_1(t, A0, attenuation, w)))
    Er_array_11.append(err)
    

    
```

```python
plt.plot(h_arr, Er_array_11, label="Зависимость ошибки от h")
plt.xlabel('h')
plt.ylabel('Error')
plt.legend()
plt.title('Первая производная с первым порядком точности')
plt.grid()
```

```python
plt.plot(np.log(h_arr), np.log(Er_array_11), label="Зависимость ошибки от h в логарифмическом масштабе")
plt.xlabel('ln(h)')
plt.ylabel('ln(Error)')
plt.legend()
plt.title('Первая производная с первым порядком точности')
plt.grid()
```

## Первая производная со вторым порядком точности

Функция dif_12(func_cell, t) рассчитывает первую производную сеточной функции (func_cell) на отрезке t(list t) cо вторым порядком точности

```python
def dif_12(func_cell, t):
    dif_12 = [0] * len(t)
    dif_12[0] = (-3 * func_cell[0] + 4 * func_cell[1] - func_cell[2]) / (t[2] - t[0])
    dif_12[-1] = (func_cell[-3] - 4 * func_cell[-2] + 3 * func_cell[-1]) / (t[-1] - t[-3])
     
    for i in range (1, len(func_cell) - 1):
        dif_12[i] = (func_cell[i+1] - func_cell[i-1]) / (t[i+1] - t[i-1])
        
    return dif_12
    
```

```python
N_min = 10
N_max = 1000
Er_array_12 = []
h_arr_12 = []
for N in range (N_min, N_max):
    t = np.linspace(0, period, N)
    h = period / (N - 1)
    h_arr_12.append(h)
    func_cell = func(t, A0, attenuation, w)
    D_12 = dif_12(func_cell, t)
    err = max(np.abs(D_12 - real_dif_1(t, A0, attenuation, w)))
    Er_array_12.append(err)
    
```

```python
plt.plot(h_arr_12, Er_array_12, label="Зависимость ошибки от h")
plt.xlabel('h')
plt.ylabel('Error')
plt.legend()
plt.title('Первая производная со вторым порядком точности')
plt.grid()
```

```python
plt.plot(np.log(h_arr_12), np.log(Er_array_12), label="Зависимость ошибки от h в логарифмическом масштабе")
plt.xlabel('ln(h)')
plt.ylabel('ln(Error)')
plt.legend()
plt.title('Первая производная со вторым порядком точности')
plt.grid()
```

## Вторая производная

Функция "real_dif_2" будет рассчитывать вторую производную по точной формуле.

```python
def real_dif_2(t, A0, att, w):
    return A0 * np.exp(-t * att) * ((att * att - w * w) * np.sin(w * t) - 2 * w * att * np.cos(w * t))
```

Функция dif_22(func_cell, t) рассчитывает вторую производную сеточной функции (func_cell) на отрезке t(list t) cо вторым порядком точности.

```python
def dif_22(func_cell, t):
    dif_22 = [0] * len(func_cell)
    h = t[1] - t[0]
    dif_22[0] = (2 * func_cell[0] - 5 * func_cell[1] + 4 * func_cell[2] - func_cell[3]) / (h * h)
    dif_22[-1] = (2 * func_cell[-1] - 5 * func_cell[-2] + 4 * func_cell[-3] - func_cell[-4]) / (h * h)
    for i in range(1, len(func_cell) - 1):
        dif_22[i] = (func_cell[i+1] - 2 * func_cell[i] + func_cell[i-1]) / (h * h)
    return dif_22
```

```python
t1 = np.linspace(0, period, 12)
t = np.linspace(0, period, 100)
f = func(t1, A0, attenuation, w)
D22 = dif_22(f, t1)
plt.plot(t1, D22, label = 'расчетная вторая производная')
plt.plot(t, real_dif_2(t, A0, attenuation, w), label='точная вторая производная')
plt.legend()
```

```python
N_min = 10
N_max = 1000
Er_array_22 = []
h_arr_22 = []
for N in range (N_min, N_max):
    t = np.linspace(0, period, N)
    h = period / (N - 1)
    h_arr_22.append(h)
    func_cell = func(t, A0, attenuation, w)
    D_22 = dif_22(func_cell, t)
    err = max(np.abs(D_22 - real_dif_2(t, A0, attenuation, w)))
    Er_array_22.append(err)
```

```python
plt.plot(h_arr_22, Er_array_22, label="Зависимость ошибки от h")
plt.xlabel('h')
plt.ylabel('Error')
plt.legend()
plt.title('Вторая производная со вторым порядком точности')
plt.grid()
```

```python
plt.plot(np.log(h_arr_22), np.log(Er_array_22), label="Зависимость ошибки от h в логарифмическом масштабе")
plt.xlabel('ln(h)')
plt.ylabel('ln(Error)')
plt.legend()
plt.title('Вторая производная со вторым порядком точности')
plt.grid()
```
